## Tratamento dos dados
Quanto aos **a.** dados faltantes, que constam como '?' nesse dataset, decidi não fazer nada a princípio, porque entendo que os dados faltantes não devem atrapalhar a análise exploratória dos dados. Toda a análise abaixo levará em conta que o resultado pode ser afetado pela ausência desses dados. Entretanto, os dados faltantes de NativeCountry foram convertidos, como explicado abaixo.

Assim, para a presente análise, o dataset exige o tratamento quanto:
 - Aos **b.** dados duplicados;
 - À **c.** reestruturação das colunas **MaritalStatus**, **RelationshipStatus**, **NativeCountry** e **Income**; e 
 - À **d.** mudança de `dtypes` das variáveis categóricas.

```python
# copy the dataset before cleaning and continue the cleaning and analysis in df
adult = df.copy()
```

#### b. Dados Duplicados

```python
df.drop_duplicates(inplace=True)
assert df.duplicated().sum() == 0 # no output if true
```

#### c. Reestruturação de colunas 
##### MaritalStatus
**MaritalStatus** conta com vários valores para explicar uma classificação binária, *married* ou *not_married*. Portanto, o atributo será reestruturado na coluna **IsMarried** para contar com apenas valores True ou False, tendo em vista que ter mais de duas categorias para explicar apenas uma classificação não parece relevante.

```python
df['IsMarried'] = df['MaritalStatus'].copy()

df['IsMarried'] = df['IsMarried'].mask(((df['IsMarried']=='Never-married')), other=False)
df['IsMarried'] = df['IsMarried'].mask(((df['IsMarried']=='Widowed')), other=False)
df['IsMarried'] = df['IsMarried'].mask(((df['IsMarried']=='Divorced')), other=False)
df['IsMarried'] = df['IsMarried'].mask(((df['IsMarried']=='Separated')), other=False)
df['IsMarried'] = df['IsMarried'].mask(((df['IsMarried']=='Married-spouse-absent')), other=False)

df['IsMarried'] = df['IsMarried'].where(((df['IsMarried']==False)), other=True)
```

```python
df['IsMarried'].unique()
```

```python
del df['MaritalStatus']
```

##### Relationship
**Relationship** conta com vários valores diferentes que explicam diferentes qualidades. Como o mais relevante que se pode extrair desse atributo, que não está em MaritalStatus, é *Own a child*, defini a coluna **OwnsChild** que conta com valores True ou False para quem tem, ou não, filho.

```python
df['OwnsChild'] = df['Relationship'].copy()
df['OwnsChild'] = df['OwnsChild'].mask(df['OwnsChild'] == 'Own-child', other=True)
df['OwnsChild'] = df['OwnsChild'].where(df['OwnsChild']==True, other=False)
```

```python
df['OwnsChild'].unique()
```

```python
del df['Relationship']
```

##### NativeCountry
Como explicado, **NativeCountry** conta com várias categorias pobremente representadas dentro do dataset e existe mais valor em analisar todo o atributo de forma binária, entre *US native* e *Not US native*. Assim, criei a coluna **USNative**, que conta com os valores True ou False para quem é, ou não, nativo dos Estados Unidos. Como a grande maioria das pessoas é nativa dos Estados Unidos, e após verificar que os dados faltantes poderiam ser representativos, decidi substituir os dados faltantes por True.

```python
df['USNative'] = df['NativeCountry'].copy()
df['USNative'] = df['USNative'].mask((df['USNative']=='United-States'), other=True)
df['USNative'] = df['USNative'].where(df['USNative']==True, other=False)
```

```python
df['USNative'].unique()
```

```python
del df['NativeCountry']
```

##### Income
**Income** conta com uma classificação binária para valores maiores que USD50.000 ou menores que USD50.000. Defini a coluna **HighIncome** que conta com valores True ou False para quem tem, ou não, uma renda maior que USD50.000.

```python
df['HighIncome'] = df['Income'].copy()
df['HighIncome'] = df['HighIncome'].mask(df['HighIncome'] == '<=50K', other=False)
df['HighIncome'] = df['HighIncome'].where(df['HighIncome']==False, other=True)
```

```python
df['HighIncome'].unique()
```

```python
del df['Income']
```

#### Mudar os `dtypes`

```python
from pandas.api.types import CategoricalDtype

# define a function to do the job
def astype_cat(col, ordered=False):
    categories = df[col].unique()
    categories = pd.api.types.CategoricalDtype(categories, ordered=ordered)
    df.loc[:, col] = df.loc[:, col].astype(categories)
```

```python
cat_columns = ['Workclass', 'Occupation', 
               'Race', 'Gender']

for col in cat_columns:
    astype_cat(col)

# use ordered_education defined above to set ordered categorical dtypes to Education
categories = CategoricalDtype(ordered_education, ordered=True)
df.loc[:, 'Education'] = df.loc[:, 'Education'].astype(categories)
```

```python
df.to_csv('adult_clean.csv', index=False)
```

#### Generalização dos Gráficos
Para facilitar a visualização dos dados categóricos, defini a função abaixo para traçar a distribuição de cada um.

```python
def plot_dist_cat(col, dfcat=dfcat):
    ax = sb.countplot(data=dfcat, y=col, color=sb.color_palette('muted', 9)[0])
    total = dfcat.shape[0]
    for loc, value in zip(ax.get_yticks(), dfcat.groupby(col).size()):
        ax.text(x=value, y=loc, s=f'{100*value/total: .2f}%', va='center')
    plt.title(f"{col} Distribution", fontsize=13)
    plt.grid(lw=0, axis='y', ls='-')
    sb.despine(top=True);
```

[1] Segue abaixo a descrição dos atributos deste dataset retirada de https://www.kaggle.com/wenruliu/adult-income-dataset e complementada por mim após a análise do dataset.

1. **Age**: idades das pessoas representadas no dataset -- atributo quantitativo discreto.
2. **Workclass**: setor de trabalho em que a pessoa está inserida -- atributo categórico que conta com 17 categorias: Private, Self-emp-not-inc, Self-emp-inc, Federal-gov, Local-gov, State-gov, Without-pay, Never-worked.
3. **FinalWeight**: FinalWeight corresponde a uma classificação dada pelo Census Bureau dos Estados Unidos, organismo que proveu os dados -- atributo quantitativo discreto.
4. **Education**: nível de educação da pessoa representada -- atributo categórico que conta com as categorias Bachelors, Some-college, 11th, HS-grad, Prof-school, Assoc-acdm, Assoc-voc, 9th, 7th-8th, 12th, Masters, 1st-4th, 10th, Doctorate, 5th-6th, Preschool.
5. **EducationNum**: nível de educação da pessoa representada -- atributo categórico com representação numérica ordenada referente a cada uma das categorias em Education.
6. **IsMarried**: estado civil -- atributo categórico binário.
7. **Occupation**: ocupação em que a pessoa representada se enquadra, não o seu status de trabalho -- atributo categórico que conta com 14 categorias: Tech-support, Craft-repair, Other-service, Sales, Exec-managerial, Prof-specialty, Handlers-cleaners, Machine-op-inspct, Adm-clerical, Farming-fishing, Transport-moving, Priv-house-serv, Protective-serv, Armed-Forces.
8. **OwnsChild**: tem, ou não, filhos - atributo categórico binário.
9. **Race**: cor/etnia da pessoa representada -- atributo categórico que conta com 5 categorias: White, Asian-Pac-Islander, Amer-Indian-Eskimo, Other, Black.
10. **Gender**: sexo da pessoa representada -- atributo categórico binário.
11. **CapitalGain**: ganhos de capital excluídos os ganhos de salário -- atributo quantitativo discreto.
12. **CapitalLoss**: perdas de capital excluídos os gastos mensais  -- atributo quantitativo discreto.
13. **HoursPerWeek**: horas de trabalho por semana -- atributo quantitativo discreto.
14. **USNative**: nativo, ou não, dos Estados Unidos -- atributo categórico binário.
15. **HighIncome**: possui, ou não, renda maior que 50.000 USD / ano  -- atributo categórico binário.
