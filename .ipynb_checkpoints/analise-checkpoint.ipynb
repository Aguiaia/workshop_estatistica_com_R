```python
#Importando as bibliotecas necessárias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

#Leitura do dataset
incm = pd.read_csv('adult.csv')
```

Para ter uma visão geral dos dados, usarei os métodos `.head()`, `.info()` e `.describe()`.

```python
incm.head()
```

O objetivo é ver como certos fatores influenciam na renda. Assim, a coluna *income* é a target, enquanto

    - age
    - educational-num
    - hours-per-week
    - workclass
    - education
    - marital-status
    - occupation
    - relationship
    - race
    - gender
    - native country
    
são os fatores que podem influenciar a target, ou seja, as features. Assim, essas serão as colunas utilizadas principalmente na análise.

Nota-se entradas com "?" em algumas colunas categóricas, o que indica que o entrevistado provavelmente não quis fornecer a informação ou ela não foi recolhida por algum outro motivo.

A descrição das variáveis de interesse são:
    
    - Age: idade; numérica.
    - Workclass: modalidade de trabalho; categórica: Private, Self-emp-not-inc, Self-emp-inc, Federal-gov, Local-gov, State-gov, Without-pay, Never-worked.
    - Education: maior nível de formação concluído; categórica: Bachelors, Some-college, 11th, HS-grad, Prof-school, Assoc-acdm, Assoc-voc, 9th, 7th-8th, 12th, Masters, 1st-4th, 10th, Doctorate, 5th-6th, Preschool.
    - Education-num: maior nível de formação concluído em escala númerica.
    - Marital-status: estado civil; categórica: Married-civ-spouse, Divorced, Never-married, Separated, Widowed, Married-spouse-absent, Married-AF-spouse.
    - Occupation: ocupação em termos gerais; categórica: Tech-support, Craft-repair, Other-service, Sales, Exec-managerial, Prof-specialty, Handlers-cleaners, Machine-op-inspct, Adm-clerical, Farming-fishing, Transport-moving, Priv-house-serv, Protective-serv, Armed-Forces.
    - Relationship: principal relação do indivíduo com outras pessoas; categórica: Wife, Own-child, Husband, Not-in-family, Other-relative, Unmarried.
    - Race: grupo étnico; categórica: White, Asian-Pac-Islander, Amer-Indian-Eskimo, Other, Black.
    - Gender: gênero; categórica: Male, Female.
    - Hours-per-week: jornada de trabalho semanal; numérica.
    - Native-country: país de origem; categórica: United-States, Cambodia, England, Puerto-Rico, Canada, Germany, Outlying-US(Guam-USVI-etc), India, Japan, Greece, South, China, Cuba, Iran, Honduras, Philippines, Italy, Poland, Jamaica, Vietnam, Mexico, Portugal, Ireland, France, Dominican-Republic, Laos, Ecuador, Taiwan, Haiti, Columbia, Hungary, Guatemala, Nicaragua, Scotland, Thailand, Yugoslavia, El-Salvador, Trinadad&Tobago, Peru, Hong, Holand-Netherlands.
    - Income: renda anual; categórica: <=50k, >50k.
 
 
As categorias de *education* podem ser resumidas.

Já *native-country* apresenta categorias demais, o que dificulta a visualização dos dados. Por ser um censo dos EUA, a vasta maioria dos indivíduos deve ser nativo do país, por isso creio ser mais eficaz dividir a variável entre *'United-States'* e *'Other'* (territórios não-incorporados dos EUA estarão na última). *"South"* e *"Hong"* são categorias em que houve algum erro de digitação, provavelmente correspondem a "South Korea" e "Hong Kong" e creio que não há problemas em juntá-las a 'Other'.

```python
educ = ['11th', '10th', '9th', '7th-8th', '12th', '1st-4th', 'Preschool', '5th-6th']

for cat in educ:
    incm['education'] = incm['education'].replace(cat,'basic-ed-inc')
    
natco = ['Cambodia', 'England', 'Puerto-Rico', 'Canada', 'Germany', 'Outlying-US(Guam-USVI-etc)', 'India',
         'Japan', 'Greece', 'China', 'Cuba', 'Iran', 'Honduras', 'Philippines', 'Italy', 'Poland', 'Jamaica',
         'Vietnam', 'Mexico', 'Portugal', 'Ireland', 'France', 'Dominican-Republic', 'Laos', 'Ecuador', 'Taiwan',
         'Haiti', 'Columbia', 'Hungary', 'Guatemala', 'Nicaragua', 'Scotland', 'Thailand', 'Yugoslavia', 'El-Salvador',
         'Trinadad&Tobago', 'Peru', 'Holand-Netherlands', 'South', 'Hong']

for cat in natco:
    incm['native-country'] = incm['native-country'].replace(cat,'Other')
```

```python
incm.info()
```

Como todas as colunas estão marcadas com *non-null*, não há entradas nulas, o que atrapalharia na manipulação dos dados.

```python
incm.describe()
```

Algo que chama atenção é que tanto para *capital-gain* quanto *capital-loss* o valor de pelo menos 75% das observações é 0, ou seja, a maioria das pessoas apresentam balança de capital neutra. No entanto, tais colunas não serão exploradas profundamente.

# Análise Numérica

Aqui serão analisadas *age*, *educational-num* e *hours-per-week* e sua relação.

### Age

```python
incm['age'].hist()
plt.show()
```

Apesar de os valores de *age* irem de 17 a 90 anos, pelo histograma vesmo que a concentração maior dos dados é de 20 a 45 (mediana: 37) e o número de observações cai a partir de 50, o que reflete os dados da pirâmide etária estadunidense.

### Educational num 

```python
incm['educational-num'].hist()
plt.show()
```

Os dados estão concentrados em torno da média (10.078) e mediana (10) e variam pouco ($\mathrm{std} = 2.57$, $\dfrac{\mathrm{std}}{\mathrm{mean}} = 0.255$).
Apesar da maioria significativa dos valores estarem entre 9 e 12 e caírem abruptamente a partir daí, há um pico em 14.

### Hours per week

```python
incm['hours-per-week'].hist()
plt.show()
```

Aqui a média e a mediana também coincidem (aproximadamente 40 horas) e há pouca variação ($\mathrm{std} = 12.39$, $\dfrac{\mathrm{std}}{\mathrm{mean}} = 0.3$), ou seja, os dados se concentram fortemente em torno da média.

## Relação entre as features

```python
#criar um dataframe novo apenas com as colunas de interesse
incmnum = incm[['age', 'educational-num', 'hours-per-week', 'capital-gain','capital-loss', 'fnlwgt']].copy()

plt.figure(figsize=(7, 4))
sns.heatmap(incmnum.corr(), vmin=-1, vmax=1, annot=True, fmt='.2f')
plt.show()
```

Como *income* é uma variável categórica, não é possível incluí-la no heatmap. Mesmo assim, com ele é possível analisar a correlação das features entre si. Observa-se, porém, que elas tem quase nenhuma correlação.

Além disso, verifica-se que *capital-gain*, *capital-loss*, *final weight* não são muito úteis para a análise.

```python
incmnum.plot(x='age', y='hours-per-week', alpha=.05, kind='scatter')
plt.show()

incmnum.plot(x='age', y='educational-num', alpha=.05, kind='scatter')
plt.show()

incmnum.plot(x='hours-per-week', y='educational-num', alpha=.05, kind='scatter')
plt.show()
```

Os scatterplots confirmam a falta de correlação. 

As conclusões que se pode tirar deles são esperadas. No primeiro, vemos que a maioria das observações de menos de 40h de trabalho semanais tem menos de 25 anos ou mais de 60. No segundo, vemos que o grau máximo de educação aos 20 anos é 10 e esse máximo vai aumentando até os 30 anos; ou seja, as pessoas só atinjem o grau 16 a partir dos 30.

## Conclusões

Fatores que correspondem a maior probabilidade de ter renda anual superior a 50 mil são:

    - Maiores jornadas de trabalho semanais;
    - Maiores graus de educação (especialmente doutorado e graduações classificadas como "professional school");
    - Idade próxima a 40 anos;
    - Trabalho no setor público federal ou como dono de empresa própria;
    - Estado civil: ser casado (provavelmente pela divisão de despesas);
    - Naturalidade: ser nascido nos EUA;
    - Gênero: ser do gênero masculino;
    - Etnia: ser branco ou asiático.
    
Os três últimos provavelmente estão ligados à questão da desigualdade social, já que identificou-se que para mulheres, imigrantes, negros e outros grupos que não brancos ou asiáticos, a proporção de pessoas com baixos graus educacionais é maior.

Dentre as features, as que mostram afetar mais tal probabilidade são:

    - age;
    - educational-num/education;
    - hours-per-week;
    - gender;
    - race;
    - native-country.
